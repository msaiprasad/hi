```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#imported the csv file.
df1=pd.read_csv("stroke.csv")
```

```python
#Retriving the first '10' records.
df1.head(10)
```

```python
#counting how many patient had a stroke or not.
df1['stroke'].value_counts()
```

```python
# 1. BMI column having NaN Values. That should be replaced with the mean of BMI Values.
# It will show the records of patients suffered with stroke but having the NaN of BMI Attribute

df1[df1['bmi'].isna() & df1['stroke']==1]
```

```python
# Replacing the NaN values with mean of BMI Attribute
import numpy as np
```

```python
df1['bmi'].fillna(np.round(df1['bmi'].mean(),1),inplace=True)
```

```python
df1
```

```python
df1['smoking_status'].value_counts()
```

```python
df1['gender'].value_counts()
```

```python
#That 'other 1 gender' needs to be removed because size is insignificant.
df2=df1[df1['gender']!='Other']
```

```python
df2
```

```python
#Drop the 'id' column
df2.drop(columns=['id'],inplace=True)
```

```python
df2
```

```python
#Data types of all columns
df2.dtypes
```

```python
#For correlation the Attributes should be 'float' and 'integers'
df3=df2.select_dtypes(include=['float64','int'])
df3.head()
```

```python
#creating a correlation 
correlation=df3.corr()
correlation
```

## Data Analysis

```python
#Plotting a heatmap
import seaborn as sns

plt.figure(figsize=(10,6))

sns.heatmap(correlation, cmap='YlGnBu',annot=True)
plt.show()
```

## ML Model

```python
#Label Encoding
col=df2.select_dtypes(include=['object']).columns
print(col)
```

```python
from sklearn.preprocessing import LabelEncoder
# Initializing our Label Encoder object
Le=LabelEncoder()
# Transfer categorical data into numeric
df2[col]=df2[col].apply(Le.fit_transform)
print(df2.head(10))
```

```python
df2
```

```python
#Find correlation and plot heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df2.corr(),cmap='YlGnBu',annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

#Splitting of Data
train_x,test_x,train_y,test_y=train_test_split(df2[col],df2['stroke'],test_size=0.25)

#Shape of the Data
train_x.shape,test_x.shape,train_y.shape,test_y.shape
```
